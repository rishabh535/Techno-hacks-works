```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\srish\OneDrive\Desktop\Iris.csv")

```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
categories = df['Species'].unique()
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
colors = ['red', 'blue', 'green', 'purple']  

for i, feature in enumerate(features):
    values = [df[df['Species'] == category][feature].mean() for category in categories]
    plt.bar(categories, values, color=colors[i], label=feature)

plt.xlabel('Species')
plt.ylabel('Mean Values')
plt.title('Bar Chart of Data Distribution')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
